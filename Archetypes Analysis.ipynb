```python
import pandas as pd

import chargen
```

```python
archs = chargen.read_archetypes()
```

```python
df = pd.DataFrame(
    [a.abilities for a in archs],
    index=[a.name for a in archs])
df.fillna("2d", inplace=True)
df["traits"] = [a.traits for a in archs]
df["training"] = [a.training for a in archs]
df["resources"] = [a.resources for a in archs]
df["techniques"] = [a.techniques for a in archs]
df["specialty"] = [a.specialty for a in archs]
df.head()
```
